```python
%load_ext autoreload
%autoreload 2

import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from helper import Helper
```

```python
fit_results_path = '../model_fit_results/tta_and_d/'
exp_data = pd.read_csv('../data/measures.csv')
exp_data.loc[(exp_data.d_condition==90) & (exp_data.tta_condition==4), 'RT'] = np.nan
```

```python
for model in os.listdir(fit_results_path):
    model_fit_results_path = os.path.join(fit_results_path, model)    
    fig, axes = Helper.plot_var_by_subject(data=exp_data, fit_results_path=model_fit_results_path, 
                               var='is_turn_decision', ylabel='probability of turn')
    fig.suptitle('Model %s' % model[0], fontsize=30)
    Helper.plot_var_by_subject(data=exp_data[exp_data.is_turn_decision], fit_results_path=model_fit_results_path, 
                               var='RT', ylabel='response time')
    model_rts = pd.read_csv(os.path.join(model_fit_results_path, 'rt_pdf.csv'))
    Helper.plot_rt_pdfs(exp_data, model_rts)
```
